```python
for dataset in data:
    num_params = len(dataset)
    
    for param in dataset:
        file.write(param['label'] + " (" + param['unit'] + ")\t")
        
    file.write("\n")
    
    for i in range(len(dataset[0]['data'])):
        for param in dataset:
            file.write(str(param['data'][i]) + "\t")
        file.write("\n")
```
